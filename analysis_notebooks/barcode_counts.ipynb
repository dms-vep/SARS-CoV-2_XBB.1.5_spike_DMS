```python
#read in count data
Lib1_493C_1 = pd.read_csv('../results/antibody_escape/by_selection/Lib1-230720-sera-493C_highACE2/Lib1-230720-sera-493C_0.16_highACE2_prob_escape.csv')
Lib1_493C_1 = Lib1_493C_1[["barcode",
                         "aa_substitutions",
                         "prob_escape",
                         "antibody_count",
                         "no-antibody_count"
                      ]]
Lib1_493C_1['antibody_concentration'] = '0.16'
Lib1_493C_2 = pd.read_csv('../results/antibody_escape/by_selection/Lib1-230720-sera-493C_highACE2/Lib1-230720-sera-493C_0.045_highACE2_prob_escape.csv')
Lib1_493C_2 = Lib1_493C_2[["barcode",
                         "aa_substitutions",
                         "prob_escape",
                         "antibody_count",
                         "no-antibody_count"
                      ]]
Lib1_493C_2['antibody_concentration'] = '0.045'

Lib1_493C_3 = pd.read_csv('../results/antibody_escape/by_selection/Lib1-230720-sera-493C_highACE2/Lib1-230720-sera-493C_0.091_highACE2_prob_escape.csv')
Lib1_493C_3 = Lib1_493C_3[["barcode",
                         "aa_substitutions",
                         "prob_escape",
                         "antibody_count",
                         "no-antibody_count"
                      ]]
Lib1_493C_3['antibody_concentration'] = '0.091'
```

```python
#read in count data
Lib1_493C = pd.concat([Lib1_493C_1, Lib1_493C_2, Lib1_493C_3], ignore_index=True)
```

```python
Lib1_493C= Lib1_493C.loc[Lib1_493C['no-antibody_count'] > 20]

Lib1_493C
```

## N234D

```python
#plot only retained
N234D = Lib1_493C[Lib1_493C["aa_substitutions"].str.contains("N234D")==True]

barcode = alt.selection_single(
    on="mouseover", fields=["barcode"], empty="none",
)
    
corr_chart = (
    alt.Chart(N234D)
    .encode(
        x=alt.X(
            "no-antibody_count",
            title="no-antibody_count",
            axis=alt.Axis(grid=False),
            scale=alt.Scale(domainMax=3000, domainMin = 0),
        ),
        y=alt.Y(
            "antibody_count",
            title="antibody_count",
            axis=alt.Axis(grid=False),
            scale=alt.Scale(domainMax=3000, domainMin = 0),
        ),
        column='antibody_concentration',
        # color = 'prob_escape:Q',
        tooltip=[
            alt.Tooltip(c, format=".3g") if N234D[c].dtypes == float
            else c
            for c in N234D.columns.tolist()
        ],
        opacity=alt.condition(barcode, alt.value(1), alt.value(0.8)),
        fill=alt.Fill(
                "prob_escape:Q",
                title="escape probability",
                legend=alt.Legend(direction="horizontal"),
                scale=alt.Scale(scheme="yellowgreenblue")),
        size=alt.condition(barcode, alt.value(120), alt.value(60)),
        strokeWidth=alt.condition(barcode, alt.value(1.5), alt.value(0)),
    )
    .mark_circle(filled=True, size=60, opacity=0.1)
    .configure_axis(grid=False)
    .properties(width=200, height=200)
    .add_selection(barcode)
)
corr_chart
```
