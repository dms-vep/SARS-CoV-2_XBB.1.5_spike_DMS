```python
import functools
import itertools
import math
import operator
import os

import altair as alt

import pandas as pd

_ = alt.data_transformers.disable_max_rows()
```

## Read the data sets

```python
xbb_spike_func_Lib1_csv = pd.read_csv(xbb_spike_func_Lib1_csv).assign(
    experiment="XBB.1.5 full spike lentivirus",
    library="Lib1")
xbb_spike_func_Lib1_csv = xbb_spike_func_Lib1_csv.loc[xbb_spike_func_Lib1_csv['times_seen'] >= times_seen]
xbb_spike_func_Lib1_csv
```

```python
xbb_spike_func_Lib2_csv = pd.read_csv(xbb_spike_func_Lib2_csv).assign(
    experiment="XBB.1.5 full spike lentivirus",
    library="Lib2")
xbb_spike_func_Lib2_csv = xbb_spike_func_Lib2_csv.loc[xbb_spike_func_Lib2_csv['times_seen'] >= times_seen]
xbb_spike_func_Lib2_csv
```

```python


XBB_func = pd.merge(xbb_spike_func_Lib1_csv, xbb_spike_func_Lib2_csv, on=['wildtype', 'site', 'mutant'])
XBB_func
```

```python
XBB_func['mutation']= XBB_func['wildtype']+XBB_func['site'].astype(str)+XBB_func['mutant']
XBB_func['times_seen'] =  XBB_func[['times_seen_x','times_seen_y']].mean(axis=1)
```

```python
XBB_func = XBB_func[["mutation",
                         "functional_effect_x",
                         "functional_effect_y",
                         "times_seen"
                           ]]
XBB_func
```

```python
slider_1 = alt.binding_range(
    min=XBB_func["times_seen"].min(),
    max=XBB_func["times_seen"].max(),
    step=1,
    name="times seen",
)    
selector_1 = alt.selection_point(
    name="SelectorName1",
    fields=["times_seen"],
    bind=slider_1,
    value= 3,
)


corr_base = (
    alt.Chart(XBB_func)
    .add_params(selector_1)
    .transform_filter(alt.datum["times_seen"] >= times_seen)
)



corr_chart = (
    (
        (
            corr_base
            .encode(
                x=alt.X("functional_effect_x",title="Lib-1 functional score", scale=alt.Scale(nice=False, padding=3)),
                y=alt.Y("functional_effect_y", title="Lib-2 functional score", scale=alt.Scale(nice=False, padding=3)),
                tooltip=XBB_func.columns.tolist(),
            )
            .mark_circle(color="black", size=60, opacity=0.15)
        )
        + (
            corr_base
            .transform_regression("functional_effect_x", "functional_effect_y", params=True)
            .transform_calculate(
                r=alt.expr.if_(
                    alt.datum["coef"][1] > 0,
                    alt.expr.sqrt(alt.datum["rSquared"]),
                    -alt.expr.sqrt(alt.datum["rSquared"]),
                ),
                r_text="r = " + alt.expr.format(alt.datum["r"], ".2f"),
            )
            .encode(
                text="r_text:N",
                x=alt.value(15),
                y=alt.value(15),
            )
            .mark_text(size=20, align="left", color="blue")
        )
    )
    .properties(
        width=300,
        height=300,
    )
    .configure_axis(grid=False, labelFontSize=12, titleFontSize=16)
    .properties(
        title=alt.TitleParams(
            "XBB.1.5 biological replicate functional score correlation",
            anchor="middle",
            fontSize=16,
        ),
    )
    .add_params(selector_1)
    .transform_filter(alt.datum["times_seen"] > selector_1["times_seen"])

)
corr_chart
```
